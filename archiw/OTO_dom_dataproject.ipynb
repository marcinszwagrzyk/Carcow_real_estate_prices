```python
# dane sciagniete skryptem: https://ofedoruk.pl/szukamy-mieszkania-za-pomoca-pythona-i-beautiful-soup/
folder = r"c:\Users\marci\git\OTODOM"
os.chdir(folder)
# czytanie pliku - ignoruje zle linie
df = pd.read_csv('offers.csv', header = 0, sep = ";", error_bad_lines=False)
```

```python
# data cleaning
lista_kolumn = ['opis', 'cena', 'cena_za_metr', 'metraz']

df[lista_kolumn] = df[lista_kolumn].replace([' ', '/', 'zł', 'm²'],'', regex=True)
df[lista_kolumn] = df[lista_kolumn].replace(',','.', regex=True)
lista_float = ['cena_za_metr', 'cena', 'metraz']

#konwersja typu danej klumny
df[lista_float] = df[lista_float].astype('float64')
# usuwanie duplikatow
df1 = df.drop_duplicates(['opis', 'cena', 'cena_za_metr'], keep='first')

# outliery z common sense, selekcja poniekad z lekcji 19
df1 = df1.loc[df1['cena_za_metr'] < 25000]
df1 = df1.loc[df1['cena_za_metr'] > 3000]

# tu zajmiemy sie datami, robimy date time kolumne i robimy ja indekse, nastepnie robimy agregacje
# wyciecie daty ze stringu pelnego formatu czasu
df1["data"] = df1["data"].str.split(" ")
df1['data'] = df['data'].astype('str')
df1["data"] = df1["data"].str.split(" ", n = 1, expand = True)
df1['data'] = df1['data'].replace('-','', regex=True)

# konwersja na format daty
df1['data'] = pd.to_datetime(df1['data'], format ='%Y%m%d')
df1.set_index('data', inplace = True)
df_weekly = df1.resample('W').mean()
#print(df_weekly)

df_weekly2 = df_weekly.loc[df_weekly['cena_za_metr'] > 8000]
print(df_weekly2)

# wykres sredniej cceny
fig, ax = plt.subplots()
fig.set_size_inches(12, 4)
df_weekly2.cena_za_metr.plot(ax = ax)
```

```python
#distplot - na raz kernel oraz histogram
#sns.distplot(df2['cena_za_metr'], bins =25)
df1['dzielnica']= df1['dzielnica'].replace('Mieszkanie na sprzedaż: ','', regex=True)
#wybrac kilka dzielnic do boxplot - kazimierz, stare miasto, krowodrza, nowa huta. POSORTOWAC na wykresie?
lista_dzielnic = ['Kazimierz', "Nowa Huta", "Podgórze", "Krowodrza","Bronowice", "Ruczaj", "Kliny-Zacisze", 'Grzegórzki']
df2 = df1.loc[df1['dzielnica'].isin(lista_dzielnic)]

df2.head()
fig = df2.boxplot(column=['cena_za_metr'], by='dzielnica', figsize=[12,8])
```
